```python
import pandas as pd
import numpy as np
```

## **데이터 불러오기**

```python
df = pd.read_csv('./data/temp_new2.csv')
df.head()
```

```python
df.info()
```

```python
# 중복 데이터 제거
df2 = df.drop_duplicates().reset_index(drop = True)
df2.head()
```

## **데이터 전처리**

```python
df2['REG_DATE'] = pd.to_datetime(df2['REG_DATE'])
df2['YEAR'] = df2['REG_DATE'].dt.year
df2['MONTH'] = df2['REG_DATE'].dt.month
df2['DAY'] = df2['REG_DATE'].dt.day
df2.head()
```

```python
# x,y 컬럼 분할
y = df2['Translated_genre_name']
x = df2.drop(['Translated_genre_name','REG_DATE','Male','Female'], 1)
```

```python
# 타입별로 컬럼 지정
COL_DEL=['USER_ID_hash']
COL_DATE=['YEAR','MONTH','DAY']
COL_NUM=['PURCHASE_FLG','DISCOUNT_PRICE']
COL_CAT=['SEX_ID','Tr_Pref_Name','Tr_small_area_name']
```

```python
# 데이터 분할
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42, stratify = y)
x_tr, x_val, y_tr, y_val = train_test_split(x_train[COL_NUM + COL_CAT + COL_DATE], y_train, test_size = 0.2, random_state = 42, stratify = y_train)
```

```python
# 스케일링
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_tr[COL_NUM] = scaler.fit_transform(x_tr[COL_NUM])
x_val[COL_NUM] = scaler.transform(x_val[COL_NUM])
x_test[COL_NUM] = scaler.transform(x_test[COL_NUM])
```

```python
# 인코딩
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

X = pd.concat([x_train[COL_CAT], x_test[COL_CAT]])

for col in COL_CAT:
  le.fit(X[col])
  x_tr[col] = le.transform(x_tr[col])
  x_val[col] = le.transform(x_val[col])
  x_test[col] = le.transform(x_test[col])
```

```python
x_tr.info()
```

## **1. XGBoost**

```python
y_tr
```

```python
# y값 인코딩 > xgboost는 y값이 문자형이면 실행되지 않아서 라벨 인코딩 진행
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()

y_tr_ = label.fit_transform(y_tr)
y_val_ = label.transform(y_val)
y_test_ = label.transform(y_test)
```

```python
np.unique(y_tr)
```

```python
np.unique(y_tr_)
```

```python
from xgboost import XGBClassifier

# ---------- 하이퍼 파라미터 종류 ----------
# n_estimators : 학습 모델의 수, 많아질수록 성능 향상의 가능성이 있으나, 속도가 느려짐 (default 100)
# learning_rate : 학습률, 너무 크면 gradient 발산의 가능성이 있으며, 너무 작으면 학습이 느림 (default 0.1, 일반적으로 0.01 ~ 0.2 사용(너무 작게 하면 시간 더 오래 걸림))
# max_depth : 최대 탐색 깊이, 너무 크면 과적합의 가능성, 너무 작으면 학습 성능 저하 (default 6, 일반적으로 3~10 사용(값 크게할수록 과적합 감소))
# min_samples_split : 분할 종료 최소 샘플 수, 큰 수면 과적합을 막지만 학습 성능 저하 가능성
# min_samples_leaf : leaf node가 되기 위한 최소 샘플 수, min_samples_split과 비슷한 용도
# random_state : 특정 숫자로 고정 시에 실행 시마다 고정된 결과 반환
# ------------------------------------------

# 모델 선언
xgb_clf = XGBClassifier(random_state = 42, objective = 'multi:softmax', num_class = 13)

# 모델 학습
xgb_clf.fit(x_tr, y_tr_)
```

```python
from sklearn.metrics import accuracy_score, recall_score, classification_report

# 예측
xgb_pred = xgb_clf.predict(x_val)

# 정확도/재현율 확인
xgb_accuracy = accuracy_score(y_val_, xgb_pred)

# 클래스별 분류 보고서 생성
xgb_report = classification_report(y_val_, xgb_pred)

print('xgb_accuracy :', xgb_accuracy) #0.8092~8114
print('xgb_report :\n', xgb_report)
```

```python
# 변수 중요도 확인 및 시각화
xgb_clf.feature_importances_
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

xgb_ftr_importances = xgb_clf.feature_importances_
xgb_ftr_importances = pd.Series(xgb_ftr_importances, index = x_tr.columns)
xgb_ftr_importances_sort = xgb_ftr_importances.sort_values(ascending = False)

sns.barplot(x = xgb_ftr_importances_sort, y = xgb_ftr_importances.index, palette = 'Set2')
```

```python
# xgb 시각화 - 오버샘플링 후
from xgboost import to_graphviz
import graphviz

to_graphviz(xgb_clf)
```

```python
# 오버샘플링
from imblearn.over_sampling import SMOTE

# 오버샘플링 객체 생성
smote = SMOTE(random_state = 42)

# 샘플링 적용
x_tr_over, y_tr_over = smote.fit_resample(x_tr, y_tr_)
print('SMOTE 적용 전 학습용 피처/레이블 데이터 셋 :', x_tr.shape, y_tr_.shape)
print('SMOTE 적용 후 학습용 피처/레이블 데이터 셋 :', x_tr_over.shape, y_tr_over.shape)
print('SMOTE 적용 후 전 분포 :\n', pd.Series(y_tr_).value_counts())
print('SMOTE 적용 후 값 분포 :\n', pd.Series(y_tr_over).value_counts())
```

```python
# 오버샘플링 후

from xgboost import XGBClassifier

# 모델 선언
xgb_clf_over = XGBClassifier(random_state = 42, objective = 'multi:softmax', num_class = 13)

# 모델 학습
xgb_clf_over.fit(x_tr_over, y_tr_over)
```

```python
# 오버샘플링 후

from sklearn.metrics import accuracy_score, recall_score, classification_report

# 예측
xgb_pred_over = xgb_clf_over.predict(x_val)

# 정확도/재현율 확인
xgb_accuracy = accuracy_score(y_val_, xgb_pred_over)

# 클래스별 분류 보고서 생성
xgb_report = classification_report(y_val_, xgb_pred_over)

print('xgb_accuracy :', xgb_accuracy)
print('xgb_report :\n', xgb_report)
```

```python
# 오버샘플링 후
import xgboost
import matplotlib.pyplot as plt

xgboost.plot_importance(xgb_clf_over, xlabel = '')
```

```python
# 오버샘플링 후
import matplotlib.pyplot as plt
import seaborn as sns

xgb_ftr_importances = xgb_clf_over.feature_importances_
xgb_ftr_importances = pd.Series(xgb_ftr_importances, index = x_tr.columns)
xgb_ftr_importances_sort = xgb_ftr_importances.sort_values(ascending = False)

sns.barplot(x = xgb_ftr_importances_sort, y = xgb_ftr_importances.index, palette = 'Set2')
```

```python
# xgb 시각화 - 오버샘플링 후
from xgboost import to_graphviz
import graphviz

to_graphviz(xgb_clf_over)
```
